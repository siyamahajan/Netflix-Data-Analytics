```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
```

```python
# Load dataset
data = pd.read_csv('netflix1.csv')
print(data.head())
```

```python
# Data Cleaning
data.drop_duplicates(inplace=True)
```

```python
data.dropna(subset=['director', 'country'], inplace=True)
```

```python
# EDA
plt.figure(figsize=(8, 6))
sns.barplot(x=data['type'].value_counts().index, y=data['type'].value_counts().values, palette='Set2')
plt.title('Distribution of Content by Type')
plt.xlabel('Type')
plt.ylabel('Count')
plt.savefig('content_distribution.png')  # Save the figure
plt.show()
```

```python
# Most Common Genres
data['genres'] = data['listed_in'].apply(lambda x: x.split(', '))
all_genres = sum(data['genres'], [])
genre_counts = pd.Series(all_genres).value_counts().head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x=genre_counts.values, y=genre_counts.index, palette='Set3')
plt.title('Most Common Genres on Netflix')
plt.xlabel('Count')
plt.ylabel('Genre')
plt.savefig('common_genres.png')  # Save the figure
plt.show()
```

```python
# Content Added Over Time
data['date_added'] = pd.to_datetime(data['date_added'], errors='coerce')  # Convert with error handling
data['year_added'] = data['date_added'].dt.year  # Extract the year

plt.figure(figsize=(12, 6))
sns.countplot(x='year_added', data=data, palette='coolwarm')
plt.title('Content Added Over Time')
plt.xlabel('Year')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.savefig('content_added_over_time.png')  # Save the figure
plt.show()
```

```python
# Top 10 Directors
top_directors = data['director'].value_counts().head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x=top_directors.values, y=top_directors.index, palette='Blues_d')
plt.title('Top 10 Directors with the Most Titles')
plt.xlabel('Number of Titles')
plt.ylabel('Director')
plt.savefig('top_directors.png')  # Save the figure
plt.show()
```

```python
# Word Cloud for Movie Titles
movie_titles = data[data['type'] == 'Movie']['title']
wordcloud = WordCloud(width=800, height=400, background_color='black').generate(' '.join(movie_titles))

plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud of Movie Titles')
plt.savefig('word_cloud.png')  # Save the figure
plt.show()
```
